Chatbot Intelligence and Logic

```python
# Define the conversation pairs (Support Flows & FAQs)
pairs = [
    [r"hi|hello|hey", ["Hello! Welcome to our support center. How can I assist you today?"]],
    [r"track my order|order status", ["I can help with that. Please provide your Order ID for tracking."]],
    [r"refund|return policy", ["Items can be returned within 30 days of purchase. Would you like the link to the full policy?"]],
    [r"human|agent|talk to someone", ["I am connecting you to a live representative. Please wait..."]],
    [r"quit|bye", ["Thank you for reaching out. Have a great day!"]],

    # Smart Fallback Handling: This handles unrecognized queries
    [r"(.*)", ["I'm sorry, I didn't quite understand that. Could you try rephrasing your question?"]]
]
```

Web Deployment

```python
import gradio as gr

def respond(message, history):
    # Initialize the chat with our defined pairs
    chat = Chat(pairs, reflections)
    return chat.respond(message)

# Launch with share=True to generate a public link for your demo
gr.ChatInterface(fn=respond, title="Customer Support Assistant").launch(share=True)
```
